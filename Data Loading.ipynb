```python
# actual code to load
images = []
age = []
gender = []
ethnicity = []
i = 0

# load first index images
dir_list = os.listdir(data_dir)
for image in dir_list:
    img = cv2.imread(data_dir + '/' + image)
    labels = image.split("_")
    if(len(image.split("_")) < 3):
        continue
    age.append(labels[0])
    gender.append(labels[1])
    ethnicity.append(labels[2])
    images.append(img)
    i = i + 1
```
